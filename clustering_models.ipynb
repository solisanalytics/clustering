# Clustering Algorithms: Code Walkthrough Using Scikit-Learn

### Import Modules/Packages

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import (
    silhouette_score, 
    davies_bouldin_score, 
    calinski_harabasz_score
)
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd

plt.style.use('fivethirtyeight')
```

### Sample Dataset

```python
data = np.array([
    [1, 2], [1.5, 1.8], [5, 8],
    [8, 8], [1, 0.6], [9, 11]]
)

print(data)
```

### Clustering Models

```python
# applying K-Means
kmeans = KMeans(n_clusters=2, max_iter=300, tol=1e-4)
kmeans.fit(data)
clusters = kmeans.predict(data)
centroids = kmeans.cluster_centers_
```

```python
# visualizing kmeans clusters
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1],
            c='red', s=200, alpha=0.5)
plt.title('K-Means Clustering')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# applying DBSCAN
dbscan = DBSCAN(eps=3, min_samples=2).fit(data)
labels = dbscan.labels_
print(labels)
```

```python
# visualizing DBSCAN
plt.scatter(data[:, 0], data[:, 1], 
            c=labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# Applying Hierarchical Clustering
linked = linkage(data, 'single')
print(linked)
```

```python
# Dendrogram Visualization
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Hierarchical Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

### Evaluating Clustering Metrics - KMeans

```python
# Silhouette Coefficient
score = silhouette_score(data, kmeans.labels_)
print(f'Silhouette Coefficient: {score:.2f}')
```

```python
# Elbow Method
wcss = []

for i in range(1,6):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
# Elbow Method Plot
plt.plot(range(1, 6), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Davies-Bouldin Index
db_index = davies_bouldin_score(data, kmeans.labels_)
print(f'Davies-Bouldin Index: {db_index:.2f}')
```
